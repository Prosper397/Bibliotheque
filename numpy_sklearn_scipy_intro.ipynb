```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

X = load_iris().data  # 4 features
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

print(X_reduced[:5])
```

```python
import networkx as nx

G = nx.Graph()
edges = [ (1, 2, 1), (2, 3, 2), (1, 3, 5)]
G.add_weighted_edges_from(edges)

path = nx.shortest_path(G, source=1, target=3, weight='weight')
print("Chemin le plus court :", path)
```

```python
from scipy.optimize import linprog

# Minimiser  3x + 4y  sous contraintes :
# x + 2y >= 8
# x >= 0, y >= 0

result = linprog(c=[3,4], A_ub=[[-1,-2]], b_ub=[-8], bounds=(0, None))
print(result)
```
